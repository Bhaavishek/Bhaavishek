# Data Analysis and Visualization of the Cleaned data

```python
# Check the percentage of cancelled reservations
cancelled_perc= df['is_canceled'].value_counts(normalize = True)
cancelled_perc
```

```python
# Based on the analysis 37% are cancelled which is very high ( ideal cancellation would be less than 10%)

reservation_perc = df['is_canceled'].value_counts()
plt.figure(figsize =(5,5))
plt.title('Resevation status')
plt.bar(['Not cancelled' , 'cancelled'], reservation_perc , edgecolor = 'k' , width = 0.5)
plt.show()
```

```python
# Now checking which hotel  type cancellation rate is higher

plt.figure(figsize=(8,4))
ax1 = sns.countplot(x = 'hotel' , hue = 'is_canceled', data = df, palette = 'Blues')
plt.title('Reservation Status in different hotels', size = 15)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

```python
# checking resort hotel and city hotel cancellation %

resort_hotel = df[df['hotel'] == 'Resort Hotel']

resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']

city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Checking date wise average daily rate

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate of City and Resort Hotel', fontsize = 20)
plt.plot(resort_hotel.index , resort_hotel['adr'] , label ='Resort Hotel')
plt.plot(city_hotel.index , city_hotel['adr'] , label ='Resort Hotel')
plt.show()
```

```python
# we need to check average dailty rate based on Month
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x = 'month' , hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation Status per month' , fontsize = 15)
plt.xlabel('month')
plt.ylabel('Number of Reservation')
plt.legend(['Not canceled' , 'Canceled'])
plt.show()
```

```python
#Plot Avg daily rate for each month based on canceled data

plt.figure(figsize=(15,8))
plt.title("Avg. daily rate per month", fontsize = 15)
sns.barplot('month','adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
# Cancellation rate based on Countries ( Top 10 country out of 177)
cancelled_data = df[df['is_canceled'] == 1]

top10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (20,8))
plt.title('Top 10 country with Cancellation rate' , fontsize=15)
plt.pie(top10_country  , autopct = '%.2f', labels = top10_country.index )
plt.show()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace= True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

notcancelled_data = df[df['is_canceled'] == 0]
notcancelled_df_adr = notcancelled_data.groupby('reservation_status_date')[['adr']].mean()
notcancelled_df_adr.reset_index(inplace= True)
notcancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize=(20,6))
plt.title('Average daily Rate' , fontsize = 20)
plt.plot(notcancelled_df_adr['reservation_status_date'] , notcancelled_df_adr['adr'], label = 'Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'] , cancelled_df_adr['adr'], label = 'Cancelled')
plt.legend()
```
